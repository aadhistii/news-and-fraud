## Functions and Methods

```python
def remove_emojis(text):
  text = re.sub(r'[\U00010000-\U0010ffff]', '', text)

  return text
```

```python
def remove_link(text):
  if isinstance(text, str):
      url_pattern = r'https?://\S+|www\.\S+|http?://\S+'
      cleaned_text = re.sub(url_pattern, '', text)
      return cleaned_text.strip()
  else:
      return text  # Return unchanged if not a string
```

```python
def remove_username(text):
  # Remove username mentions starting with '@'
  text = re.sub(r'@\w+', '', text)

  # Remove username mentions starting with 'RT @' for retweets
  text = re.sub(r'RT @\w+:?', '', text)

  # Remove username mentions enclosed in square brackets for Reddit-style mentions
  text = re.sub(r'\[\/?u\/\w+\]', '', text)

  return text
```

```python
def remove_punctuation(text):
    if isinstance(text, str):
        return ''.join([char for char in text if char not in string.punctuation])
    else:
        return text
```

```python
# Function to extract text before and after (separate 1 text into 2 text)

def extract_text(text, word):
  pattern = re.compile(re.escape(word), re.IGNORECASE)
  parts = pattern.split(text, 1)
  before_word = parts[0].strip()
  after_word = parts[1].strip() if len(parts) == 2 else None
  return before_word if before_word else None, after_word if after_word else None
```

```python
def extract_text_after(text, word, df=None, column_label=None):
    # Check if a DataFrame and a column label are provided
    if df is not None and column_label is not None:
        # Apply the function only to rows where the specified column has a certain value (e.g., 1)
        text = text[df[column_label] == 1]

    # Regular expression pattern with case-insensitive flag
    pattern = re.compile(re.escape(word), re.IGNORECASE)
    parts = pattern.split(text, 1)
    after_word = parts[1] if len(parts) == 2 else None
    return after_word.strip() if after_word else None
```

```python
def extract_links_after_word(text, word):
    url_pattern = r'https?://\S+|www\.\S+|http?://\S+'
    matches = re.findall(word + r'\s+(' + url_pattern + r')', text.lower())
    return matches
```

```python
def extract_text_after_quote(text, word):
    pattern = re.escape(word) + r'\s*\"(.*?)\"'
    match = re.search(pattern, text)
    if match:
        return match.group(1).strip()
    return None
```

```python
df_clean = df.applymap(lambda x: remove_emojis(x) if isinstance(x, str) else x)
df_clean = df_clean.applymap(lambda x: remove_username(x) if isinstance(x, str) else x)
df_clean[['pre_narasi', 'post_narasi']] = df_clean['Body'].apply(lambda x: extract_text(x, 'narasi')).apply(pd.Series)
df_clean['hoax'] = df_clean['Body'].apply(lambda x: extract_text_after(x, 'hoax:')).apply(pd.Series)
df_clean['hoax'] = df_clean['hoax'].apply(lambda x: remove_link(x)).apply(pd.Series)
df_clean['fakta'] = df_clean['Body'].apply(lambda x: extract_text_after(x, 'fakta:')).apply(pd.Series)
df_clean['fakta'] = df_clean['fakta'].apply(lambda x: remove_link(x)).apply(pd.Series)
df_clean['hoax_url'] = df['Body'].apply(lambda x: extract_links_after_word(x, 'hoax:'))
```
